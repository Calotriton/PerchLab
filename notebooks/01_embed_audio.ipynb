# 01 — Embedding Generation

Generate Perch V2 embeddings and store them in a Perch Hoplite database.
This mirrors `perchlab embed` / the interactive **Embedding Generation** workflow.


## Configure
Set the input folder (optionally one subfolder per species when `labeled=True`).

```python
from pathlib import Path
from perchlab import configure_logging
from perchlab.config import AppConfig
from perchlab.workflows.embed import EmbeddingWorkflow

configure_logging('INFO')
config = AppConfig()
config.embed.input_dir = Path('recordings')   # <-- edit
config.embed.output_dir = Path('embeddings_out')
config.embed.labeled = False
config.embed.export = 'parquet'
```

## Run

```python
summary = EmbeddingWorkflow().run(config)
print('processed:', summary.processed, 'embeddings:', summary.detections)
```

## Inspect the Hoplite database

```python
from perch_hoplite.db import sqlite_usearch_impl
db = sqlite_usearch_impl.SQLiteUSearchDB.create(db_path=str(config.embed.output_dir / 'hoplite_db'))
print('embeddings stored:', db.count_embeddings())
```
