# 03 — Benchmark

Evaluate Perch V2 on a labelled dataset (one subfolder per species). Mirrors
`perchlab benchmark`. Produces metrics, a confusion matrix, ROC/PR curves, and a
Markdown report.


```python
from pathlib import Path
from perchlab import configure_logging
from perchlab.config import AppConfig
from perchlab.workflows.benchmark import BenchmarkWorkflow

configure_logging('INFO')
config = AppConfig()
config.benchmark.input_dir = Path('labelled_dataset')  # <-- edit
config.benchmark.output_dir = Path('benchmark_out')
config.benchmark.sweep.enabled = True
config.benchmark.sweep.start, config.benchmark.sweep.end, config.benchmark.sweep.step = 0.0, 1.0, 0.1
```

## Run

```python
summary = BenchmarkWorkflow().run(config)
```

## View results

```python
import json
print(json.dumps(json.loads((config.benchmark.output_dir / 'metrics.json').read_text()), indent=2)[:800])
```
