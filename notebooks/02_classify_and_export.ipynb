# 02 — Species Identification

Run Perch V2's classifier over recordings, keep top-k per window, apply a
confidence threshold, and export CSV + Raven selection tables. Mirrors
`perchlab id`.


```python
from pathlib import Path
from perchlab import configure_logging
from perchlab.config import AppConfig
from perchlab.workflows.identify import SpeciesIdentificationWorkflow

configure_logging('INFO')
config = AppConfig()
config.identify.input_dir = Path('recordings')   # <-- edit
config.identify.output_dir = Path('perchID')
config.identify.top_k = 3
config.identify.threshold = 0.6
```

## Run

```python
summary = SpeciesIdentificationWorkflow().run(config)
print('detections:', summary.detections)
```

## Load the aggregate table

```python
import pandas as pd
df = pd.read_csv(config.identify.output_dir / 'threshold_0.60' / 'all_detections.csv')
df.head()
```
